```python
import random
answer = random.sample(range(1,10),4)
print(f'答案{answer}')
a=b=n=0
num=0
while a!=4:
  num +=1
  a=b=n=0
  user= input('輸入四個數字(不會重複):')
  if len(user)>len(answer):
    user= input('輸入數字過多，輸入四個數字(不會重複):')
  elif len(user)<len(answer):
    user= input('輸入數字過少，輸入四個數字(不會重複):')
  for i in user:
    if int(user[n])==answer[n]:
     a+=1
    else:
      if int(i) in answer:
        b+=1
    n+=1
  output=','.join(user).replace(',','')
  print(f'{output}: {a}A{b}B')
print(f'答對了!總共猜了{num}次')
```
