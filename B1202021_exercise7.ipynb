```python
def tree(h):
    for i in range(1, h + 1):
        a = " " * (h - i)
        b = "*" * (2 * i - 1)
        print(a+b)
    c = " " * (h - 2)
    d = "***"
    for _ in range(5):
        print(c+d)
h = int(input("聖誕樹層數: "))
tree(h)
```

```python
def bmi(height, weight):
    bmi = weight / ((height / 100) ** 2)
    return bmi

def output(bmi):
    if bmi < 18.5:
        return "過輕"
    elif bmi < 24:
        return "適中"
    elif bmi < 27:
        return "過重"
    else:
        return "肥胖"

try:
    height = float(input("身高（公分）: "))
    weight = float(input("體重: "))
    bmi = bmi(height, weight)
    print("BMI:", bmi)
    print("體位:", output(bmi))
except:
    print("輸入內容錯誤")
```
