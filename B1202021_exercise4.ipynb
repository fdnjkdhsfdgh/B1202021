```python
a=float(input('請輸入體重(kg):'))
b=float(input('請輸入身高(cm):'))
c=(a/((b/100))**2)
print('BMI:',c)
if c<18.5:
     print('體位:過輕')
elif c>=18.5 and c<24:
     print('體位:適中')
elif c>=24 and c<27:
     print('體位:過重')
elif c>=27:
     print('體位:肥胖')
```

```python
a=float(input('請輸入第一個數字:'))
b=(input('請輸入運算符號:'))
c=float(input('請輸入第二個數字:'))
if b=='+':
  print(a+c)
elif b=='-':
  print(a-c)
elif b=='*':
  print(a*c)
elif b=='/':
  print(a/c)
```
